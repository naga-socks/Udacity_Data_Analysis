# Check Dataset 1
- df.shape -- matrix size
- df.head -- look at first portion of data
- df.tail -- look at last portion of data
- df.info -- summary of dataframe
- df.describe -- useful statistics for each column of data
- df.nunique -- returns number of unique values in each column
- isnull().sum() -- sum number of missing features
- duplicated().sum() -- sum number of duplicated values

```python
print(color.BOLD + color.UNDERLINE + '\n-----DATASET 1-----\n' + color.END)
print(a_df.nunique())
print('\nThe unique values for dataset 1 --> ', a_df.Fuel.unique())

print(color.BOLD + color.UNDERLINE + '\n-----DATASET 2-----\n' + color.END)
print('\nThe unique values for dataset 2 --> ', b_df.Fuel.unique())
```
