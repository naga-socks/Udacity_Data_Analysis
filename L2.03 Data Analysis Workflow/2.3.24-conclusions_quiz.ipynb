```python
# imports and load data
import pandas as pd

%matplotlib inline

df = pd.read_csv('store_data.csv')
df.head()
```

```python
# explore data
df.info();

#change week "object" to "date"
df['week'] = pd.to_datetime(df['week']);
```

```python

# check data to see if it changed
df.info();
```

```python
# compares and creates scatter plots and histogram of every 
pd.plotting.scatter_matrix(df, figsize=(15,15));

# plot box plots for each variable
df.plot.box(figsize=(10,10));
```

```python
# explore data
df.hist(figsize=(8, 8));

# this returns useful descriptive statistics for each column of data
df.describe()
```

```python
#see the last n rows
df.tail(20)
```

```python
# total sales for the last month
df.iloc[196:, 1:].sum()
```

```python
# average sales
df.mean()
```
